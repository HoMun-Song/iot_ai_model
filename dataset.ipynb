```python
import cv2
import numpy as np
from glob import glob
from tqdm import tqdm
import os
```

```python
path = '/home/workspace/iot_ai_model/dataset/supervisely_person/images_bak/*.png'
files = glob(path)
```

```python
size = []
for f in tqdm(files):
    img = cv2.imread(f)
    if np.min(img.shape[:2]) > 512:
        ratio = img.shape[0] / img.shape[1]
        if ratio < 1 : # h < w
            img = cv2.resize(img, (int(512 / ratio), 512))
        else :
            img = cv2.resize(img, (512, int(512 * ratio)))
        filename = os.path.basename(f)
        
    cv2.imwrite('/home/workspace/iot_ai_model/dataset/supervisely_person/images/' +filename, img)
```
